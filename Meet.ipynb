```python
from stanza.models.constituency import tree_reader, transition_sequence
from collections import defaultdict
import re
import stanza
from summarizer import Summarizer
import logging
from summarizer.sbert import SBertSummarizer
import spacy_fastlang
import pandas as pd
import numpy as np
from transformers import pipeline

stanza.download('en')
#visit_preorder

class query_creation():
  nlp_constituency_parser = stanza.Pipeline('en', processors='tokenize,pos,constituency', package={'constituency': 'wsj_bert'})
  classifier = pipeline("summarization")
  def __init__(self, text, summary=None):
    self.text = text
    self.query_list = []
    self.extractive_summarizer = Summarizer()
    self.summary = summary
  
  
  def get_extractive_summary(self, raw_unsummarized_text):
    summarized_text = self.extractive_summarizer(raw_unsummarized_text)
    return summarized_text
  
  @staticmethod
  def get_abstractive_summary(raw_unsummarized_text):
    summarized_text = query_creation.classifier(raw_unsummarized_text)[0]['summary_text']
    return summarized_text

  def get_summary(self):
    if self.summary=='extractive':
      return self.get_extractive_summary(self.text)
    elif self.summary=='abstractive':
      return query_creation.get_abstractive_summary(self.text)
    else:
      logging.info("No summarizer setected")
      return self.text

  def get_query_for_text(self):
    sentence = self.get_summary()
    #query_creation.get_abstractive_summary(self.text)
    doc = query_creation.nlp_constituency_parser(sentence)
    
    for sentence in doc.sentences:
      #print(sentence.constituency)
      depth, children = query_creation.get_first_branch(sentence.constituency)
      sentence_query = []
      for child in children:
        sentence_query.append(query_creation.get_proto_query(child, flatten=True))
      #print(sentence_query)
      sentence_query = query_creation.remove_empty_string(sentence_query)
      sentence_query = ['('+_+')' for _ in sentence_query] 
      self.query_list.append(' AND '.join(sentence_query))
    
    #  print(sentence.constituency.visit_preorder(print, print, print))
    #  print(list(transition_sequence.build_sequence(sentence.constituency)))
  
  @staticmethod
  def get_first_branch(tree_object, depth=0):
    children = tree_object.children
    if tree_object.is_leaf():
      return None
    if len(children)>1:
      return depth, children
    return query_creation.get_first_branch(tree_object.children[0], depth+1)
  
  
  @staticmethod
  def are_children_preterminal(tree_object):
    return all([child.is_preterminal() for child in tree_object.children])
  
  @staticmethod
  def get_children_by_terminality(tree_object):
    preterminal_str = 'preterminal'
    leaf_str = 'leaf'
    internal_str = 'internal'
    terminality_dict = defaultdict(list)
    for child in tree_object.children:
      if child.is_preterminal():
        terminality_dict[preterminal_str].append(child)
      elif child.is_leaf():
        terminality_dict[leaf_str].append(child)
      else:
        terminality_dict[internal_str].append(child)
    return terminality_dict
  
  @staticmethod
  def get_flattened_NPs(tree_object):
    return ','.join(tree_object.leaf_labels())
  
  @staticmethod
  def get_proto_query(tree_object, flatten=True):
    if flatten:
      proto_query = query_creation.get_flattened_tree(tree_object)
      processed_proto_query = re.sub(r'[,]+', ',',re.sub(r'(\sOR\s)+', ' OR ', proto_query))
      if processed_proto_query[:4]==' OR ':
        processed_proto_query = processed_proto_query[4:]
      print(processed_proto_query)
      return re.sub(r'(\sOR\s,\sOR\s)+', ', ', processed_proto_query)

  @staticmethod
  def remove_empty_string(list_of_strings):
    return [_ for _ in list_of_strings if _!='']

  @staticmethod
  def get_flattened_tree(tree_object, flag=0):
    preterminal_str = 'preterminal'
    leaf_str = 'leaf'
    internal_str = 'internal'
    #breakpoint()
    # leaf_labels is a method that gets all the leaf lables of a tree. Just use it to get all the NP results
    if tree_object.label=='NP' or flag:
      # if all the children are pre-terminal
      if query_creation.are_children_preterminal(tree_object):
        #print("Pre_terminal: ", tree_object.label, tree_object)
        temp_return_list = [child.children[0].label.lower() for child in tree_object.children]
        temp_return_list = query_creation.remove_empty_string(temp_return_list)
        return_str = ', '.join(temp_return_list)
        #print(return_str)
        #breakpoint()
        return return_str
      # if a few of the children are preterminal and others are not
      else:
        return_str = ''
        terminality_dict = query_creation.get_children_by_terminality(tree_object)
        #return_str += ' OR ' + ' , '.join([leaf.label for leaf in terminality_dict[leaf_str]])
        #return_str += ' OR ' + ' , '.join([preterminal.children[0].label for preterminal in terminality_dict[preterminal_str]])
        temp_return_list = [query_creation.get_flattened_tree(child, flag=1) for child in terminality_dict[internal_str]]
        temp_return_list = query_creation.remove_empty_string(temp_return_list)
        return_str += ' OR ' + ' OR '.join(temp_return_list)
        #print(return_str)
        #breakpoint()
        return return_str
        #print("Not Pre_terminal: ", tree_object.label, tree_object)
    else:
      #breakpoint()
      temp_return_list = [query_creation.get_flattened_tree(tree_object_child, flag=0) for tree_object_child in tree_object.children]
      temp_return_list = query_creation.remove_empty_string(temp_return_list)
      return ' OR '.join(temp_return_list)


def test():
  '''
  test example from paper "Automatic Boolean Query Formulation for Systematic Review Literature Search", 
  by Harrisen Scells, Guido Zuccon, Bevan Koopman, and Justin Clark, in WWW ’20, April 20–24, 2020, Taipei, Taiwan
  '''
  query_creation_obj = query_creation("galactomannan detection for invasive aspergillosis in immunocompromised patients.")
  query_creation_obj.get_query_for_text()
  check = '(galactomannan, detection) AND (invasive, aspergillosis OR immunocompromised, patients)'
  assert query_creation_obj.query_list[0]==check, f'{query_creation_obj.query_list[0]} does not match {check}'
```

```python
'''
Books and Documents

Clinical Trial

Meta-Analysis

Randomized Controlled Trial

Review

Systematic Review
'''
from transformers import pipeline
import spacy
import yake
import nltk
from nltk.tokenize import sent_tokenize
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import itertools
import numpy as np
import torch
import random
import numpy as np

np.random.seed(373)
random.seed(373)
torch.manual_seed(373)

nlp_spacy = spacy.load('en_core_web_md')
nltk.download('punkt')

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

language = "en"
max_ngram_size = 3
deduplication_thresold = 0.8
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 20
kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)

def get_sentiment_calssifier():
  tokenizer = AutoTokenizer.from_pretrained("siebert/sentiment-roberta-large-english")
  model = AutoModelForSequenceClassification.from_pretrained("siebert/sentiment-roberta-large-english").to(device)
  #tokenizer = AutoTokenizer.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
  #model = AutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
  label_dict = {0:"NEGATIVE", 1:"POSITIVE"}
  #label_dict = {0:"NEGATIVE", 1:"NEGATIVE", 2:"NEGATIVE", 3:"POSITIVE", 4:"POSITIVE"}
  
  #results = model(**tokenizer(["I am here", "I am here", "We are everywhere, but there"], truncation=True, return_tensors='pt', padding=True))
  def sentiment_classifier(text_list: list):
    encoded_text = tokenizer(text_list, truncation=True, return_tensors='pt', padding=True)
    output_logits = None
    with torch.no_grad():
      encoded_text.to(device)
      output_logits = model(**encoded_text)
    output_arg = np.argmax(output_logits.logits.cpu().detach().numpy(), axis=1)
    #print(output_arg)
    return [i.item() for i in output_arg] #label_dict[output_arg]
  return sentiment_classifier

# get PERSON and DATE entities
def get_entities(text):
  entities = []

  text = nlp_spacy(text)
  for word in text.ents:
    #print(word.text,word.label_)
    if (word.label_=='PERSON') | (word.label_=='DATE'):
      entities.extend(word.text.split())
  #print("entities: "+str(entities)) #bamboo
  return [_.lower() for _ in entities]

def get_yake_keyphrases(text: str):
  '''
  pass the entire article and this function will return the keyphrases
  Inputs
    text: text of entire article
  Outputs
    list of keyphrases
  '''
  keywords = kw_extractor.extract_keywords(text)
  
  #for kw in keywords:
  #  print(kw)
  return keywords  

def not_in_entities_list(keyphrase_list, entities_list):
  #entities list is already falttened in get_entities method
  flattened_keyphrase_list = []
  for keyphrase in keyphrase_list:
    flattened_keyphrase_list.extend(keyphrase.lower().split())
  for keyphrase in flattened_keyphrase_list:
    if keyphrase in entities_list:
      return False
  return True

def get_yake_phrases_extractor():
  nlp_spacy = spacy.load('en_core_web_md')
    

  def extractor(text, filtering=True):
    doc = nlp_spacy(text)
    entities_list = get_entities(text)
    if filtering:
      #print(get_yake_keyphrases(text))
      #print(list(filter(lambda i: not_in_entities_list(i[0].split(), entities_list), get_yake_keyphrases(text))))
      #lambda i: all([_.lower() not in entities_list for _ in i[0].split()])
      entities_list = list(zip(*filter(lambda i: not_in_entities_list(i[0].split(), entities_list), get_yake_keyphrases(text))))
    else:
      entities_list = list(zip(*get_yake_keyphrases(text)))
    if len(entities_list)==0:
      return entities_list
    return entities_list[0]

  return extractor

def flatten_list(list_):
  # assuming that list is a 2D list
  return list(itertools.chain(*list_))

def get_clinical_med_entities_extractor():
  stanza.download('en', package='mimic', processors={'ner': 'i2b2'})
  stanza.download('en', package='mimic', processors={'ner': 'ncbi_disease'})
  nlp_clinical_ner = stanza.Pipeline('en', package='mimic', processors={'ner': 'i2b2'})
  nlp_disease_ner = stanza.Pipeline('en', package='mimic', processors={'ner': 'ncbi_disease'})
  #'The patient had a sore throat and was treated with Cepacol lozenges.'
  
  def extractor(text):
    doc_clinical_ner = nlp_clinical_ner(text)
    doc_disease_ner = nlp_disease_ner(text)    
    # return all entities
    clinical_entities_dict = defaultdict(list)
    disease_ner_list = list()
    disease_ner_list = [ent.text.lower() for ent in doc_disease_ner.entities if ent.type=='DISEASE']
    
    for ent in doc_clinical_ner.entities:
      if any([ent_text for ent_text in ent.text.lower().split() if ent_text not in disease_ner_list]):
        clinical_entities_dict[ent.type].append(ent.text.lower())
    clinical_entities_dict['DISEASE'] = disease_ner_list
    return clinical_entities_dict
        #print(f'{ent.text}\t{ent.type}')
  return extractor

def get_filter_query(books_docs=False, clinical_trial=True, meta_analysis=True,\
                         randomized_controlled_trial=True, review=False, systematic_review=True, **args):
  filter_query = []
  if books_docs or clinical_trial or meta_analysis or randomized_controlled_trial or review or systematic_review:
    if books_docs:
      filter_query.append("booksdocs[Filter]")
    if clinical_trial:
      filter_query.append("clinicaltrial[Filter]")
    if meta_analysis:
      filter_query.append("meta-analysis[Filter]")
    if randomized_controlled_trial:
      filter_query.append("randomizedcontrolledtrial[Filter]")
    if review:
      filter_query.append("review[Filter]")
    if systematic_review:
      filter_query.append("systematicreview[Filter]")
  return filter_query

def reverse_med_entities_dict(dict_: dict):
  reversed_dict_ = {}
  for k, v in dict_.items():
    for phrase in v:
      reversed_dict_[phrase] = k
  return reversed_dict_

#booksdocs[Filter] OR clinicaltrial[Filter] OR meta-analysis[Filter] OR 
#randomizedcontrolledtrial[Filter] OR review[Filter] OR systematicreview[Filter]
def get_med_entities_query_creator(sentence_based = True):
  med_entities_extractor = get_clinical_med_entities_extractor()
  yake_phrases_extractor = get_yake_phrases_extractor()
  
  def extend_query_parts(text, query_parts, with_yake=False, books_docs=False, clinical_trial=True, meta_analysis=True,\
                         randomized_controlled_trial=True, review=False, systematic_review=True):
    '''
    if len(query_parts)==1 or with_yake:
      keyphrases = yake_phrases_extractor(text)
      if len(keyphrases)!=0:
        query_parts.append(keyphrases[:3])
      else:
        return ['']
    '''
    if len(query_parts)<=1:
      return ['']
    query_parts.append(get_filter_query(books_docs, clinical_trial, meta_analysis, randomized_controlled_trial,\
                          review, systematic_review))
    return query_parts

  def article_med_entities_query_creator(article_text, with_yake=False, books_docs=False, clinical_trial=True, meta_analysis=True,\
                         randomized_controlled_trial=True, review=False, systematic_review=True):
    
    sentences = sent_tokenize(article_text)
    sentences = [_.lower() for _ in sentences]
    #sentence_sentiments = sentiment_classifier(sentences) # more efficient to put all the sentences through together
    med_entities = med_entities_extractor(article_text) # concept to phrase list mapping
    reversed_med_entities = reverse_med_entities_dict(med_entities) # phrase to concept mapping
    phrase_list = []
    phrase_list = list(reversed_med_entities.keys())
    query = ''
    
    for idx, sentence in enumerate(sentences):
      # phrase to concept mapping for this sentence
      #print(sentence) #bamboo
      phrases_in_sentence = dict([(phrase, reversed_med_entities[phrase]) for phrase in phrase_list if sentence.find(phrase)!=-1])
      concept_to_phrase = defaultdict(list)
      for phrase, concept in phrases_in_sentence.items():
        concept_to_phrase[concept].append(phrase)
      query_parts = []
      #breakpoint()
      for concept_key, phrase_list_value in concept_to_phrase.items():
        query_parts.append(phrase_list_value)
      query_parts = extend_query_parts(sentence, query_parts, with_yake, books_docs, clinical_trial, meta_analysis, randomized_controlled_trial,\
                                      review, systematic_review)
      
      query_string = []
      for query_part in query_parts:
        query_string.append('('+' OR '.join(query_part)+')')
      yield ' AND '.join(query_string), sentence#, sentiment_classifier([sentence])
      


  def sentence_med_entities_query_creator(text, with_yake=False, books_docs=False, clinical_trial=True, meta_analysis=True,\
                         randomized_controlled_trial=True, review=False, systematic_review=True):
    med_entities = med_entities_extractor(text)
    
    query_parts = []
    for key, value in med_entities.items():
      query_parts.append(value)
    
    query_parts = extend_query_parts(text, query_parts, with_yake, books_docs, clinical_trial, meta_analysis, randomized_controlled_trial,\
                          review, systematic_review)
    
    query_string = []
    for query in query_parts:
      query_string.append('('+' OR '.join(query)+')')
    yield ' AND '.join(query_string)#, sentiment_classifier([text])
  if sentence_based:
    return sentence_med_entities_query_creator
  else:
    return article_med_entities_query_creator
```
